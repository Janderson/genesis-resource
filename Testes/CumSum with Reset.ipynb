```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

## Method 1

```python
df = DataFrame( {"value" : np.random.randn(50000)})

df["dir"] = np.sign(df.value)

df["trend"] = np.where(abs(df['dir'] + df['dir'].shift(1))<=1,0, df['dir'])

df["trendcount"] = df.groupby((df.trend!=0).cumsum()).cumcount()

df["trendsign"] = df.trendcount * df.trend.shift(1)

```
